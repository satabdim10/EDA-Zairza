#### Import pandas and read in the Games csv file and set it to a DataFrame called gms

```python
# your code here-
import pandas as pd

gms = pd.read_csv('games.csv')
```

```python
# your code here-
rows, columns = gms.shape

rows, columns
```

#### What is the highest number of comments received for a game?

```python
# your code here-
max(gms.total_comments)
```

#### What is the name of the game with id 1500?

```python
# your code here-
gms.loc[gms.id == 1500,['name']].name
```

#### Which games have received the least number of comments?



```python
# your code here-
gms[gms.total_comments == min(gms.total_comments)][['name','total_comments']]
```

#### Which games have received the highest number of comments?¶


```python
# your code here-
gms[gms.total_comments == max(gms.total_comments)][['name','total_comments']]
```

#### How many unique games are there in the dataset?

```python
# your code here-
gms.id.nunique()
```

#### How many boardgames and boardgameexpansions are there in the dataset?

```python
# your code here-
count_boardgames = len(gms[gms.type == 'boardgame'])
count_boardhameexpansions = len(gms[gms.type == 'boardgameexpansion'])

count_boardgames, count_boardhameexpansions
```

#### What is the mean playin time for all games put together ?

```python
# your code here-
gms.playingtime.mean(axis=0)
```
